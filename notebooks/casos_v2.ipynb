```python
import matplotlib.pyplot as plt
import numpy as np

def plot_cost_vs_probability(
    cost_correct, # I(k == e)
    cost_incorrect, # I(k != e)
    N=1000
):

    fig, ax = plt.subplots()

    qe = np.arange(N) / N
    correct = cost_correct(qe)
    incorrect = cost_incorrect(qe)

    ax.plot(qe, correct, label='Cost when k == e')
    ax.plot(qe, incorrect, label='Cost when k != e')
    ax.set_xlabel('Probability q_e')
    ax.set_ylabel('Cost')
    ax.set_title('Cost vs Probability q_e')
    ax.set_xlim(0.01, 1.01)
    ax.set_xticks(np.arange(0, 1.1, 0.1))
    # ax.set_ylim(-0.01, 1.01)
    # ax.set_yticks(np.arange(0, 1.1, 0.1))
    # ax.set_aspect('equal', adjustable='box')
    ax.legend()
    ax.grid()
```

# CCAG genérico con $\tilde C = C_{01}$

Calculamos en general:
$$
C_{\gamma}(y_k,d_j) = \begin{cases}
a(\gamma)\tilde C(y_k,d_j) & \text{if } d_j \neq d_r \\
b(\gamma) & \text{if } d_j = d_r \\
\end{cases}
$$

$$
d_B(q) = \begin{cases}
\tilde d_B(q) & \text{if } s_{\tilde C}(q) \leq \frac{b(\gamma)}{a(\gamma)} \\
d_r & \text{if } s_{\tilde C}(q) > \frac{b(\gamma)}{a(\gamma)} 
\end{cases} \\
s_{\tilde C}(q)=\sum_{k=1}^K \tilde C^*(y_k,q)q_k \\
$$

$$
C_\gamma^*(y_k,q) = \begin{cases}
a(\gamma)\tilde C^*(y_k,q) & \text{if } s_{\tilde C}(q) \leq c(\gamma)=\frac{b(\gamma)}{a(\gamma)} \\
b(\gamma) & \text{if } s_{\tilde C}(q) > c(\gamma)=\frac{b(\gamma)}{a(\gamma)} 
\end{cases}
$$


Acá tenemos dos casos. Si $c(\gamma)$ es creciente, entonces:

$$
\begin{align}
C_\gamma^*(y_k,q) &= \begin{cases}
a(\gamma)\tilde C^*(y_k,q) & \text{if } c^{-1}(s_{\tilde C}(q)) \leq \gamma \\
b(\gamma) & \text{if } c^{-1}(s_{\tilde C}(q)) > \gamma 
\end{cases} \\[1em]
C^*(y_k,q) &=\int_{0}^{c^{-1}(s_{\tilde C}(q))} w(\gamma)b(\gamma) d\gamma + \tilde C^*(y_k,q)\int_{c^{-1}(s_{\tilde C}(q))}^{\gamma_M} w(\gamma)a(\gamma) d\gamma \\
\end{align}
$$


Si $c(\gamma)$ es decreciente:

$$
\begin{align}
C_\gamma^*(y_k,q) &= \begin{cases}
a(\gamma)\tilde C^*(y_k,q) & \text{if } c^{-1}(s_{\tilde C}(q)) \geq \gamma \\
b(\gamma) & \text{if } c^{-1}(s_{\tilde C}(q)) < \gamma 
\end{cases} \\[1em]
C^*(y_k,q) &=\tilde C^*(y_k,q)\int_{0}^{c^{-1}(s_{\tilde C}(q))} w(\gamma)a(\gamma) d\gamma + \int_{c^{-1}(s_{\tilde C}(q))}^{\gamma_M} w(\gamma)b(\gamma) d\gamma
\end{align}
$$


## Caso 1


>Parámetros:
>$$
>\begin{align}
>a(\gamma) = 1 & & b(\gamma) = \gamma & & w(\gamma) = \frac{1}{\gamma} & & \tilde C(y_k,d_j) = I(k \neq j) & & \gamma_M = 1 
>\end{align}
>$$

Tenemos $c(\gamma) = \frac{b(\gamma)}{a(\gamma)}=\gamma$ y $c^{-1}(s) = s$ por lo que $c(\gamma)$ es creciente. Entonces:

$$
\begin{align}
C^*(y_k,q) &=\int_{0}^{c^{-1}(s_{\tilde C}(q))} w(\gamma)b(\gamma) d\gamma + \tilde C^*(y_k,q)\int_{c^{-1}(s_{\tilde C}(q))}^{\gamma_M} w(\gamma)a(\gamma) d\gamma \\
&=\int_{0}^{s_{\tilde C}(q)} w(\gamma)\gamma d\gamma + \tilde C^*(y_k,q)\int_{s_{\tilde C}(q)}^{\gamma_M} w(\gamma) d\gamma \\
\end{align}
$$

Si elegimos $\tilde C(y_k,d_j) = I(k \neq j)$ y llamamos $q_e = \max_j q_j$ entonces $\tilde C^*(y_k,q) = I(k \neq e)$ y $s_{\tilde C}(q) = 1 - q_e$. Si además seteamos $w(\gamma)=\frac{1}{\gamma}$ y $\gamma_M=1$:
$$
\begin{align}
C^*(y_k,q) &= \int_{0}^{1-q_e} d\gamma + I(k\neq e) \int_{1-q_e}^{1} \frac{1}{\gamma}d\gamma = 1-q_e  - I(k \neq e) \log(1-q_e)
\end{align}
$$

```python
K = 1000
cost_correct = lambda qe: 1 - qe
cost_incorrect = lambda qe: 1 - qe - np.log(1-qe)
plot_cost_vs_probability(cost_correct, cost_incorrect)
```

Tenemos $c(\gamma) = \frac{b(\gamma)}{a(\gamma)}=\gamma$ y $c^{-1}(s) = s$ por lo que $c(\gamma)$ es creciente. Entonces:

$$
\begin{align}
C^*(y_k,q) &=\int_{0}^{c^{-1}(s_{\tilde C}(q))} w(\gamma)b(\gamma) d\gamma + \tilde C^*(y_k,q)\int_{c^{-1}(s_{\tilde C}(q))}^{\gamma_M} w(\gamma)a(\gamma) d\gamma \\
&=\int_{0}^{s_{\tilde C}(q)} w(\gamma)\gamma d\gamma + \tilde C^*(y_k,q)\int_{s_{\tilde C}(q)}^{\gamma_M} w(\gamma) d\gamma \\
\end{align}
$$

Si elegimos $\tilde C(y_k,d_j) = I(k \neq j)$ y llamamos $q_e = \max_j q_j$ entonces $\tilde C^*(y_k,q) = I(k \neq e)$ y $s_{\tilde C}(q) = 1 - q_e$. Si además seteamos $w(\gamma)=2$ y $\gamma_M=1$:
$$
\begin{align}
C^*(y_k,q) &= \int_{0}^{1-q_e} 2\gamma d\gamma + I(k\neq e) \int_{1-q_e}^{1} 2 d\gamma = (1-q_e)^2 + I(k \neq e) 2q_e
\end{align}
$$

```python
K = 1000
cost_correct = lambda qe: (1-qe)**2
cost_incorrect = lambda qe: (1-qe)**2 + 2 * qe
plot_cost_vs_probability(cost_correct, cost_incorrect)
```

## Caso 2

>Parámetros:
>$$
>\begin{align}
>a(\gamma) = \frac{1}{1-\gamma} & & b(\gamma) = \frac{1}{\gamma} & & w(\gamma) = 1 & & \tilde C(y_k,d_j) = I(k \neq j) & & \gamma_M = 1 
>\end{align}
>$$

Tenemos $c(\gamma) = \frac{b(\gamma)}{a(\gamma)}=\frac{1-\gamma}{\gamma}$ y $c^{-1}(s) = \frac{1}{1+s}$ por lo que $c(\gamma)$ es decreciente. Entonces:

$$
\begin{align}
C^*(y_k,q) &=\tilde C^*(y_k,q)\int_{0}^{c^{-1}(s_{\tilde C}(q))} w(\gamma)a(\gamma) d\gamma + \int_{c^{-1}(s_{\tilde C}(q))}^{\gamma_M} w(\gamma)b(\gamma) d\gamma \\
&=\tilde C^*(y_k,q)\int_{0}^{\frac{1}{1+s_{\tilde C}(q)}} \frac{w(\gamma)}{1-\gamma} d\gamma + \int_{\frac{1}{1+s_{\tilde C}(q)}}^{\gamma_M} \frac{w(\gamma)}{\gamma} d\gamma \\
\end{align}
$$

Si elegimos $\tilde C(y_k,d_j) = I(k \neq j)$ y llamamos $q_e = \max_j q_j$ entonces $\tilde C^*(y_k,q) = I(k \neq e)$ y $s_{\tilde C}(q) = 1 - q_e$. Si además seteamos $w(\gamma)=1$ y $\gamma_M=1$:
$$
\begin{align}
C^*(y_k,q) &= I(k \neq e) \int_{0}^{\frac{1}{2-q_e}} \frac{1}{1-\gamma} d\gamma + \int_{\frac{1}{2-q_e}}^{1} \frac{1}{\gamma} d\gamma = \log(2-q_e) + I(k \neq e) \log\left(\frac{2-q_e}{1-q_e}\right)
\end{align}
$$

```python
K = 1000
cost_correct = lambda qe: np.log(2 - qe)
cost_incorrect = lambda qe: 2 * np.log(2 - qe) - np.log(1 - qe)
plot_cost_vs_probability(cost_correct, cost_incorrect)
```

## Caso 3

>Parámetros:
>$$
>\begin{align}
>a(\gamma) = 1-\gamma & & b(\gamma) = \gamma & & w(\gamma) = 2 & & \tilde C(y_k,d_j) = I(k \neq j) & & \gamma_M = 1
>\end{align}
>$$

Tenemos $c(\gamma) = \frac{b(\gamma)}{a(\gamma)}=\frac{\gamma}{1-\gamma}$ y $c^{-1}(s) = \frac{s}{1+s}$ por lo que $c(\gamma)$ es creciente. Entonces:

$$
\begin{align}
C^*(y_k,q) &=\int_{0}^{\frac{s_{\tilde C}(q)}{1+s_{\tilde C}(q)}} w(\gamma)\gamma d\gamma + \tilde C^*(y_k,q)\int_{\frac{s_{\tilde C}(q)}{1+s_{\tilde C}(q)}}^{\gamma_M} w(\gamma)(1-\gamma) d\gamma \\
\end{align}
$$

Si elegimos $\tilde C(y_k,d_j) = I(k \neq j)$ y llamamos $q_e = \max_j q_j$ entonces $\tilde C^*(y_k,q) = I(k \neq e)$ y $s_{\tilde C}(q) = 1 - q_e$. Si además seteamos $w(\gamma)=2$ y $\gamma_M=1$:
$$
\begin{align}
C^*(y_k,q) &=\int_{0}^{\frac{1-q_e}{2-q_e}} 2\gamma d\gamma + I(k \neq e)\int_{\frac{1-q_e}{2-q_e}}^{1} 2(1-\gamma) d\gamma = 2 \left(\frac{1-q_e}{2-q_e}\right)^2 + I(k \neq e) \frac{2}{(2-q_e)^2}
\end{align}
$$

```python
K = 1000
cost_correct = lambda qe: 2 * ((1-qe)/(2 - qe))**2
cost_incorrect = lambda qe: 2 * (((1-qe)/(2 - qe))**2 + 1/(2 - qe)**2)
plot_cost_vs_probability(cost_correct, cost_incorrect)
```
